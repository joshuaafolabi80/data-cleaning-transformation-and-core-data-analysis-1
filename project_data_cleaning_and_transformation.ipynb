## Analyzing the top 10 nationalities, that is top 10 countries that has the highest number of players in the various clubs.

### It can be seen that most of the players in the different football clubs hail from England, followed by Germany and spain thereafter.

```python
y['NATIONALITY'].value_counts()
```

```python
nationality_counts = y['NATIONALITY'].value_counts()[0:10]
print(nationality_counts)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(13, 12))
top_nationalities = y["NATIONALITY"].value_counts().head(10)
sns.countplot(x="NATIONALITY", data=y, order=top_nationalities.index)
plt.xlabel("Count")
plt.ylabel("Nationality")
plt.title("Top 10 Nationalities")
plt.show()
```

### Analysis and visualization of the top 10 highest paid players in descending order

```python
player_salary=y[["SHORT_NAME","WAGE_EUR"]]
```

```python
player_salary.head(10)
```

```python
player_salary=player_salary.sort_values(by=["WAGE_EUR"], ascending=False)
player_salary.head(10)
```

```python
plt.figure(figsize=(14, 12))
player_salary=player_salary.sort_values(by=["WAGE_EUR"], ascending=False).head(10)
sns.barplot(x="SHORT_NAME", y="WAGE_EUR", data=player_salary)
plt.title('Top 10 Player Salaries')
plt.show()
```
